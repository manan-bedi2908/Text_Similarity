```python
from gensim import corpora, models, similarities
import jieba
```

```python
import pandas as pd
```

```python
df = pd.read_csv(r'main_news_scraped.csv', encoding='latin1')
```

```python
texts = []
```

```python
df.head(5)
```

```python
type(df['Text'][0])
```

```python
for text in df['Text']:
    if type(text) == str:
        texts.append(text)
```

```python
texts
```

```python
keyword = 'reservations to see Firefall at Yosemite this year'
```

```python
texts = [jieba.lcut(text) for text in texts]
dictionary = corpora.Dictionary(texts)
feature_cnt = len(dictionary.token2id)
corpus = [dictionary.doc2bow(text) for text in texts]
tfidf = models.TfidfModel(corpus) 
kw_vector = dictionary.doc2bow(jieba.lcut(keyword))
index = similarities.SparseMatrixSimilarity(tfidf[corpus], num_features = feature_cnt)
sim = index[tfidf[kw_vector]]
max_similar = dict()
for i in range(len(sim)):
    print('keyword is similar to text%d: %.2f' % (i + 1, sim[i]))
    max_similar[i] = sim[i]
```

```python
Keymax = max(max_similar, key=max_similar.get) 
```

```python
print("Text", Keymax+1, "is most similar to the given keyword")
```
